```python
import matplotlib.pyplot as plt
import datetime
import numpy as np
import pandas as pd
```

```python
fig = plt.figure(figsize=(12,8))
```

```python
df1 = pd.DataFrame({'date':np.array([datetime.datetime(2021, 12,i+1)for i in range(20)]),
'blogs_read':[4,6,5,8,15,13,18,6,5,3,15,14,19,21,15,19,25,24,16,26]})

df2 = pd.DataFrame({'date':np.array([datetime.datetime(2021,12,i+1)for i in range(20)]),
'blogs_unread':[1,1,2,3,3,3,4,3,2,3,4,7,5,3,2,4,3,6,1,2]})
```

```python
plt.plot(df1.date,df1.blogs_read, label = 'blogs_read',linewidth = 3)
plt.plot(df2.date,df2.blogs_unread,color = 'red', label = 'blogs_unread', linewidth = 3)

plt.title('Data')
plt.xlabel('Date')
plt.ylabel('Blogs')

plt.legend()
plt.tight_layout()

plt.show()
```

```python
data = pd.read_csv('/content/Sales - Sheet1.csv')
```

```python
df = pd.DataFrame(data)
```

```python
X = list(df.iloc[:,0])
Y = list(df.iloc[:,1])
```
